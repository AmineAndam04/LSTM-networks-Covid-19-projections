```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("/content/ND par régions.xlsx",index_col=0)
data.tail(2)
```

```python
Regions=data.columns.values
```

```python
def split_data(array,input_step,horizon):
    X,Y=[],[]
    for i in range(len(array)):
        idx_last_y=i+input_step+horizon
        if idx_last_y <= len(array):
            Y.append(array[idx_last_y-horizon:idx_last_y])
            X.append(array[i:i+input_step])
        else :
            break
    return np.array(X),np.array(Y)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def Makeprediction(NomRegion,var):
    #Prepare data
    NewCase=data[NomRegion].values
    lag=10
    horizon=5
    X,Y=split_data(NewCase,lag,horizon)
    X = X.reshape((X.shape[0], X.shape[1], 1))
    model = keras.Sequential([
    layers.LSTM(30, activation='relu',return_sequences=True, input_shape=(10, 1)),
    layers.LSTM(30,activation='relu'),
    layers.Dense(5),
  ])
    model.compile(optimizer='adam', loss='mae')
    model.fit(X,Y,epochs=70,verbose=0)
    var=np.array(var)
    var=var.astype(float)
    var=var.reshape(1,10,1)
    Y_hat=model.predict(var)
    return Y_hat
```

```python
valeurs=data.tail(10)
projections=[]
for i in range(len(Regions)):
    nom=Regions[i]
    predictors=list(valeurs[nom].values)
    pre=Makeprediction(nom,predictors)
    projections.append(pre)
    print('Fin :' + nom)
    print('-'*40)
```
